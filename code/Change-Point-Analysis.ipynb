```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ruptures as rpt
import gc
```

```python
from read_data import *
```

```python
data = process_data()
```

```python
def calculating_changepoints(data, data_set,feature,nchp,xlabel,ylabel,title):
    dat = data[data_set]["dat"]
    feature_data = dat[feature]
    algo = rpt.BottomUp(model="l1",min_size=20000).fit(feature_data.values)
    my_bkps = algo.predict(n_bkps=nchp)
    change_point_indexes = [x - 1 for x in my_bkps ]
    change_point_indexes
    df = dat['time']
    x_values = df.iloc[change_point_indexes]    
    plot_change_points(dat,x_values,data_set,feature,xlabel,ylabel,title)
    return x_values

def plot_change_points(data, result, data_set, feature,xlabel,ylabel,title):
    #result.plot()
    x_value = data['time']
    y_value = data[feature]
    #rpt.display(data.values,result)
    plt.figure(figsize=(10, 6))
    plt.plot(x_value.values, y_value.values)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(feature)
    for xc in result:
        plt.axvline(x=xc,color='red')

    plt.savefig(os.path.join("Figures", data_set, data_set + "_" + feature + "_cp.pdf"), dpi=180, bbox_inches='tight')
    plt.show()
```

```python
visual_change_points = [2,2,0,3,2,2,3]

dataset = 'V2_00001'
hdr = data[dataset]["hdr"]
for i,nchp in zip(range(len(hdr.channel_names)),visual_change_points):
        channel = hdr.channel_names[i]
        #file_name = os.path.join(channel_dir, data_set + "_" + channel + ".pdf")
        calculating_changepoints(data, dataset, channel, nchp ,hdr.horz_units, hdr.vert_units[i], channel)
```
