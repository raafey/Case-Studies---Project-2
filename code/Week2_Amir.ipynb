```python
def plot_acf_pacf(obs, title, channel, index):
    fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
    fig.suptitle('ACF and PACF Before and After for ' + title + " for " + channel)
    # Plot ACF Before
    plot_acf(data[obs]["dat"][channel][index-1000:index], ax=ax[0,0], lags=250)
    #ax[0].set_xlabel('Lag')
    ax[0,0].set_ylabel('ACF')
    #ax[0,0].set_title('Autocorrelation Function (ACF) Before and After')

    # Plot ACF After
    plot_acf(data[obs]["dat"][channel][index:1000+index], ax=ax[0,1], lags=250)
    #ax[0].set_xlabel('Lag')
   #ax[0,1].set_ylabel('ACF')
    #ax[0,1].set_title('Autocorrelation Function (ACF) After '+ title)

    
    # Plot PACF Before
    plot_pacf(data[obs]["dat"][channel][index-1000:index], ax=ax[1,0], lags=250)
    ax[1,0].set_xlabel('Lag')
    ax[1,0].set_ylabel('PACF')
    #ax[1,0].set_title('Partial Autocorrelation Function (PACF) Before and After')

    # Plot PACF After
    plot_pacf(data[obs]["dat"][channel][index:1000+index], ax=ax[1,1], lags=250)
    ax[1,1].set_xlabel('Lag')
    #ax[1,1].set_ylabel('PACF')
    #ax[1,1].set_title('Partial Autocorrelation Function (PACF) After ' + title)

    #plt.savefig(os.path.join( data + "_" + channel + "_" + title + "_acf.png"), dpi=200, bbox_inches='tight')
    for ax in fig.get_axes():
        ax.label_outer()

    plt.tight_layout()
    plt.show()
```
